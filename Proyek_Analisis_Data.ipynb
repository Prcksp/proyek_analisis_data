## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
day_df = pd.read_csv("/content/day.csv")
day_df.head()

```

```python
hour_df = pd.read_csv("/content/hour.csv")
hour_df.head()
```

### Assessing Data

**Load DataSet**

**Informasi DataSet**

```python
print("Informasi Dataset Harian:")
day_df.info()
print("\nInformasi Dataset Per Jam:")
hour_df.info()
```

**Head 5 Informasi Teratas**

```python
print("Data Harian:")
display(day_df.head())
print("Data Per Jam:")
display(hour_df.head())
```

### Cleaning Data

**Menghapus Duplikasi**

```python
day_df = day_df.drop_duplicates()
hour_df = hour_df.drop_duplicates()
```

**Nilai yang Hilang fill up with median**

```python
for col in day_df.select_dtypes(include=['number']).columns:
    day_df[col].fillna(day_df[col].median(), inplace=True)
for col in hour_df.select_dtypes(include=['number']).columns:
    hour_df[col].fillna(hour_df[col].median(), inplace=True)
```

**Check and Recheck after cleaning**

```python
print("Data Harian setelah pembersihan:")
display(day_df.info())
print("Data Per Jam setelah pembersihan:")
display(hour_df.info())
```

## Exploratory Data Analysis (EDA)

### Eksplorasi Data Statistik Deskriptif

```python
print("Statistik Data Harian:")
display(day_df.describe())
print("Statistik Data Per Jam:")
display(hour_df.describe())
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(12, 5))
sns.lineplot(x=day_df["dteday"], y=day_df["cnt"], marker="o", linestyle="-", color="b")
plt.xticks(rotation=45)
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Penyewaan")
plt.title("Tren Penyewaan Sepeda Harian dalam Rentang Seminggu")
plt.show()

plt.figure(figsize=(12, 5))
sns.boxplot(x=hour_df["hr"], y=hour_df["cnt"], palette="coolwarm")
plt.xlabel("Jam")
plt.ylabel("Jumlah Penyewaan")
plt.title("Distribusi Penyewaan Sepeda Per Jam dalam Rentang Seminggu")
plt.show()
```

**Insight:**
- **Apakah pola trennya naik atau turun?** apabila penyewaan meningkat selama waktu tertentu, bisa dikaitkn dengan perubahan faktor eksternal seperti musim dan lainnya.
- **Apakah ada pola dari faktor eksternal?** Tren penyewaan bisa dibandingkan dengan faktor eksternal seperti musim kemarau dan hujan

### Pertanyaan 2:

```python
plt.figure(figsize=(12, 5))
sns.barplot(x=day_df["weekday"], y=day_df["cnt"], palette="viridis")
plt.xlabel("Hari dalam Seminggu")
plt.ylabel("Jumlah Penyewaan")
plt.title("Jumlah Penyewaan Sepeda Berdasarkan Hari dalam Seminggu")
plt.show()

plt.figure(figsize=(12, 5))
sns.barplot(x=hour_df["hr"], y=hour_df["cnt"], hue=hour_df["weathersit"], palette="magma")
plt.xlabel("Jam")
plt.ylabel("Jumlah Penyewaan")
plt.title("Pengaruh Cuaca terhadap Penyewaan Sepeda per Jam")
plt.legend(title="Kondisi Cuaca", labels=["Cerah", "Mendung", "Hujan Ringan", "Hujan Lebat"])
plt.show()
```

**Insight:**
- **Untuk penyewaan lebih banyak apakah pada sore hari atau pada pagi hari?** apabila ada peningkatan yang signifikan pada waktu tertentu, hal ini bisa disandingkan dengan aktifitas produktif seperti berangkat dan pulang dari sekolah atau kantor.
- **Apakah ada perbedaan banyak dan tidaknya dihari kerja dan diakhir pekan?** Analisis bisa menggunakan dengan nilai peningkatan aktivitas diakhir pekan karena adanya rekreasi.

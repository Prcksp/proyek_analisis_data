### Pertanyaan 1:

- **Apakah ada pola dari faktor eksternal?** Tren penyewaan bisa dibandingkan dengan faktor eksternal seperti musim kemarau dan hujan

### Pertanyaan 2:

```python
weather_mapping = {
    1: "Cerah",
    2: "Berawan",
    3: "Hujan"
}

hour_df["weather_desc"] = hour_df["weathersit"].map(weather_mapping)

fig, axes = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle("Pengaruh Cuaca terhadap Penyewaan Sepeda", fontsize=16)

for i, (code, weather_filter) in enumerate(weather_mapping.items()):
    filtered_weather_df = hour_df[hour_df["weather_desc"] == weather_filter]

    hour_avg = filtered_weather_df.groupby('hr')['cnt'].mean()
    colors = ["blue", "orange", "red"]  # Warna untuk setiap kondisi cuaca
    sns.barplot(
        x=hour_avg.index,
        y=hour_avg.values,
        color=colors[i],
        ax=axes[i, 0]
    )
    axes[i, 0].set_title(f"Rata-rata Penyewaan Sepeda per Jam ({weather_filter})")
    axes[i, 0].set_xlabel("Jam")
    axes[i, 0].set_ylabel("Jumlah Penyewaan")
    axes[i, 0].set_xticks(range(24))
    axes[i, 0].set_xticklabels(range(24))

    palettes = ["Blues", "Oranges", "Reds"]
    sns.boxplot(
        x=filtered_weather_df["hr"],
        y=filtered_weather_df["cnt"],
        palette=palettes[i],
        ax=axes[i, 1]
    )
    axes[i, 1].set_title(f"Distribusi Penyewaan Sepeda ({weather_filter})")
    axes[i, 1].set_xlabel("Jam")
    axes[i, 1].set_ylabel("Jumlah Penyewaan")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
plt.figure(figsize=(12, 5))

hour_avg = hour_df.groupby(['weekday', 'weathersit'])['cnt'].mean().reset_index()

barplot = sns.barplot(x='weekday', y='cnt', hue='weathersit', data=hour_avg, palette='coolwarm')

plt.xlabel("Hari dalam seminggu (0=Senin, 6=Minggu)")
plt.ylabel("Rata-rata Penyewaan")
plt.title("Rata-rata Penyewaan Sepeda Berdasarkan Hari dan Faktor Cuaca")
plt.grid(axis='y', linestyle='--', alpha=0.6)

handles, labels = barplot.get_legend_handles_labels()
plt.legend(handles=handles, title="Kondisi Cuaca", labels=["Cerah", "Mendung", "Hujan", "Ekstrem"])

plt.show()
```

**Insight:**
- **Pola harian menunjukkan dua puncak utama penyewaan** di pagi hari sekitar pukul **08.00** dan sore hari sekitar pukul **17.00**. Hal ini kemungkinan besar disebabkan oleh aktivitas masyarakat seperti berangkat dan pulang kerja atau sekolah.
- **Faktor cuaca:** Jika data cuaca tersedia, bisa ditambahkan insight tentang bagaimana kondisi cuaca (cerah, hujan, atau dingin) memengaruhi jumlah penyewaan.

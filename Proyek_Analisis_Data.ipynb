## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
day_df = pd.read_csv("/content/day.csv")
day_df.head()

```

```python
hour_df = pd.read_csv("/content/hour.csv")
hour_df.head()
```

### Assessing Data

**Pengecekan Missing Value untuk Data set Hari**

```python
print("\nCek Missing Values:")
display(day_df.isnull().sum())
```

**Pengecekan Missing Value untuk Data set Jam**

```python
print("\nCek Missing Values:")
display(day_df.isnull().sum())
```

**Informasi DataSet**

```python
print("Informasi Dataset Harian:")
day_df.info()
print("\nInformasi Dataset Per Jam:")
hour_df.info()
```

**Head 5 Informasi Teratas**

```python
print("Data Harian:")
display(day_df.head())
print("Data Per Jam:")
display(hour_df.head())
```

**Cek Data Duplikat**

```python
print("\nCek Data Duplikat Harian:")
print(day_df.duplicated().sum())
print("\nCek Data Duplikat Per Jam:")
print(hour_df.duplicated().sum())
```

**Cek Konsistensi Data**

```python
day_df['dteday'] = pd.to_datetime(day_df['dteday'], errors='coerce')
print("\nFormat Data Tanggal Harian:", day_df['dteday'].dtype)
```

```python
print("\nKolom dalam Dataset Harian:")
print(day_df.columns)
print("\nKolom dalam Dataset Per Jam:")
print(hour_df.columns)
```

**Pengecekan kemungkinan bias atau kesalahan data**

```python
print("Statistik Data Harian:")
display(day_df.describe())
print("Statistik Data Per Jam:")
display(hour_df.describe())
```

### Cleaning Data

**Menghapus Duplikasi**

```python
day_df = day_df.drop_duplicates()
hour_df = hour_df.drop_duplicates()
```

**Nilai yang Hilang fill up with median**

```python
for col in day_df.select_dtypes(include=['number']).columns:
    day_df[col].fillna(day_df[col].median(), inplace=True)
for col in hour_df.select_dtypes(include=['number']).columns:
    hour_df[col].fillna(hour_df[col].median(), inplace=True)
```

**Check and Recheck after cleaning**

```python
print("Data Harian setelah pembersihan:")
display(day_df.info())
print("Data Per Jam setelah pembersihan:")
display(hour_df.info())
```

### Eksplore

#### Eksplorasi pendistribusian data

**Distribusi dari dataset Hari**

```python
plt.figure(figsize=(8, 5))
sns.histplot(day_df['cnt'], bins=20, kde=True)
plt.title("Distribusi Jumlah Sewa sepeda Harian")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.show()
```

**Distribusi dari dataset Jam**

```python
plt.figure(figsize=(8, 5))
sns.histplot(hour_df['cnt'], bins=20, kde=True)
plt.title("Distribusi Jumlah Sewa sepeda Harian berdasarkan jam")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.show()
```

#### Agregasi Data

**Agregasi data harian berdasarkan bulan**

```python
print("\nAgregasi Data Harian Berdasarkan Bulan:")
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
day_df['bulan'] = day_df['dteday'].dt.to_period('M')
month_df = day_df.groupby('bulan').agg({'cnt': ['mean', 'sum', 'count']})
display(month_df)
```

**Agregasi Data per jam berdasarkan Jam**

```python
print("\nAgregasi Data Per Jam:")
hour_grouped = hour_df.groupby('hr').agg({'cnt': ['mean', 'sum', 'count']})
display(hour_grouped)
```

#### Korelasi Data

**Korelasi Data Harian**

```python
print("\nMatriks Korelasi Data Harian:")
day_df_numeric = day_df.select_dtypes(include=['number'])
plt.figure(figsize=(10, 8))
sns.heatmap(day_df_numeric.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Heatmap Korelasi Data Harian")
plt.show()
```

**Korelasi Data per Jam**

```python
print("\nMatriks Korelasi Data perjam:")
hour_df_numeric = hour_df.select_dtypes(include=['number'])
plt.figure(figsize=(10, 8))
sns.heatmap(hour_df_numeric.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Heatmap Korelasi Data perjam")
plt.show()
```

## Visualization & Explanatory Analysis

#### Visualisasi untuk tren Data

**Visualisasi Data Harian**

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=day_df['dteday'], y=day_df['cnt'])
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Sewa Sepeda")
plt.title("Tren Jumlah Sewa Sepeda Harian")
plt.xticks(rotation=45)
plt.show()
```

**Visualisasi data per jam**

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=hour_df['hr'], y=day_df['cnt'])
plt.xlabel("Jam")
plt.ylabel("Jumlah Sewa Sepeda")
plt.title("Tren Jumlah Sewa Sepeda per jam")
plt.xticks(rotation=0)
plt.show()
```

### Pertanyaan 1:

```python
plt.figure(figsize=(10, 5))
day_avg = day_df.groupby('weekday')['cnt'].mean()
sns.barplot(x=['Min', 'Sen', 'Sel', 'Rab', 'Kam', 'Jum', 'Sab'], y=day_avg, color='grey')
plt.xlabel("Hari dalam Seminggu")
plt.ylabel("Rata-rata Jumlah Penyewaan")
plt.title("Jumlah Penyewaan Sepeda Berdasarkan Hari dalam Seminggu")
plt.show()
```

**Insight:**
- **Apakah pola trennya naik atau turun?** apabila penyewaan meningkat selama waktu tertentu, bisa dikaitkn dengan perubahan faktor eksternal seperti musim dan lainnya.
- **Apakah ada pola dari faktor eksternal?** Tren penyewaan bisa dibandingkan dengan faktor eksternal seperti musim kemarau dan hujan

### Pertanyaan 2:

```python
plt.figure(figsize=(12, 5))
hour_avg = hour_df.groupby('hr')['cnt'].mean()
sns.barplot(x=hour_avg.index, y=hour_avg.values, color='grey')
plt.xlabel("Jam")
plt.ylabel("Jumlah Penyewaan")
plt.title("Jumlah Penyewaan Sepeda Berdasarkan Jam dalam Sehari")
plt.show()
```

**Insight:**
- **Untuk penyewaan lebih banyak apakah pada sore hari atau pada pagi hari?** apabila ada peningkatan yang signifikan pada waktu tertentu, hal ini bisa disandingkan dengan aktifitas produktif seperti berangkat dan pulang dari sekolah atau kantor.
- **Apakah ada perbedaan banyak dan tidaknya dihari kerja dan diakhir pekan?** Analisis bisa menggunakan dengan nilai peningkatan aktivitas diakhir pekan karena adanya rekreasi.
